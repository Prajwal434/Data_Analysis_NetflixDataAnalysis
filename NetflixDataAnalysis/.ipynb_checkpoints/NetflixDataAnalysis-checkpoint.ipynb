# Netflix Data Analysis Project


## Step 1: Load the Dataset
In this step, we import the required libraries like Pandas for data handling, Matplotlib for visualization, and Seaborn for advanced plotting. The Netflix dataset is loaded using `pd.read_csv()` and its first few rows are displayed to understand its structure and contents.

```python
## Step 1: Load the Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("netflix1.csv")
df.head()
```

## Step 2: Data Cleaning
Cleaning data is crucial for accurate analysis. We:
- Drop missing values using `dropna()` to avoid data gaps.
- Remove duplicate records using `drop_duplicates()` to maintain data integrity.
- **Fill missing rows** with appropriate methods such as `.fillna()` for numerical data and `.ffill()` or `.bfill()` for time series data to ensure completeness.
- **Split columns** where necessary, for example, separating 'date_added' into 'year', 'month', and 'day' for better insights and analysis.

```python
## Step 2: Data Cleaning
# Checking for missing values
df.dropna(inplace=True)

# Checking for duplicates
df.drop_duplicates(inplace=True)
```


## Step 3: Exploratory Data Analysis (EDA)
EDA helps to uncover patterns and insights within the data. We visualize:

### Content Type Distribution
A bar chart displays the distribution of content types (e.g., Movies, TV Shows), helping to understand Netflix's content library.

### Top 10 Content-Producing Countries
A horizontal bar chart highlights the top 10 countries that produce the most content on Netflix.

### Genre Analysis
A bar chart visualizes the most common genres to identify popular content types among viewers.

```python
## Step 3: Exploratory Data Analysis (EDA)
### Content Type Distribution
plt.figure(figsize=(8, 5))
sns.countplot(y='type', data=df, order=df['type'].value_counts().index)
plt.title("Content Type Distribution")
plt.show()

### Top 10 content-producing countries
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', color='skyblue')
plt.title("Top 10 Content Producing Countries")
plt.show()

### Genre Analysis
plt.figure(figsize=(10, 6))
df['listed_in'].value_counts().head(10).plot(kind='bar', color='orange')
plt.title("Top 10 Genres on Netflix")
plt.show()
```

## Step 4: Insights and Observations
Based on our analysis:
- Most content on Netflix falls under TV Shows and Movies.
- The United States is the top content producer.
- Popular genres include Documentaries, Stand-Up Comedy, and International content.

```python
## Step 4: Insights and Observations
print("Insights and Observations:")
print("1. Most content on Netflix is categorized as TV Shows and Movies.")
print("2. The United States is the top content producer on Netflix.")
print("3. Popular genres include Documentaries, Stand-Up Comedy, and International content.")
```

## Step 5: Save Cleaned Data
The cleaned dataset is saved to a new CSV file using `to_csv()` to ensure the refined data is ready for future analysis.

```python
## Step 5: Save Cleaned Data
# Save cleaned data
df.to_csv("netflix_cleaned.csv", index=False)
```
